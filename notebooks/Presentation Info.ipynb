```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\olive\Documents\GitHub\MontrealFireSafetyProject\data\node_data_cleaned.csv", index_col=0)
```

# Naive

```python
df = df.loc[df['DIST_TO_FS']!=1000]
grp = df.rename({'ARR_DRT': 'Neighborhood'}, axis=1)
grp = grp.groupby(['Neighborhood']).agg({'DIST_TO_FS': ['mean']})
grp.columns = grp.columns.get_level_values(0)
grp = grp.rename({'DIST_TO_FS': 'Average Distance to Nearest Fire Station (km)'}, axis=1)
grp.sort_values('Average Distance to Nearest Fire Station (km)', ascending=False)
```
